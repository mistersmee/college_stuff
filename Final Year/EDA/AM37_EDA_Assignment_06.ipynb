```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, PowerTransformer, QuantileTransformer
```

```python
# Step 2: Perform Standardization on the variables: Age, Fare
df1=df.copy()
dfss=df1[['Age','Fare']]
dfss
```

```python
sc=StandardScaler()
sc
```

```python
dfss=sc.fit_transform(dfss)
```

```python
dfss=pd.DataFrame(data=dfss,columns=['age','fare'])
```

```python
dfss.describe()
```

```python
df1=pd.concat([df1,dfss],axis=1)
df1
```

```python
df1.drop(['Age','Fare'],axis=1,inplace=True)
df1
```

```python
# Step 3: Perform feature scaling using Min-Max Scaler on the features: Pclass, Age, SibSp, Parch, Fare
df2=df.copy()
```

```python
dfms= df2[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
ms= MinMaxScaler()
ms
```

```python
dfms= ms.fit_transform(dfms)
dfms
```

```python
dfms= pd.DataFrame(dfms, columns=['pclass', 'age', 'sibSp', 'parch', 'fare'])
dfms
```

```python
df2=pd.concat([df2,dfms],axis=1)
```

```python
df2.drop(['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'],axis=1,inplace=True)
df2
```

```python
df2.describe()
```

```python
# Step 4: Perform feature scaling using Robust Scaler on the features: Pclass, Age, SibSp, Parch, Fare
df3=df.copy()
```

```python
dfrs= df3[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
rs= RobustScaler()
rs
```

```python
dfrs= rs.fit_transform(dfrs)
dfrs
```

```python
dfrs= pd.DataFrame(dfrs, columns=['pclass', 'age', 'sibSp', 'parch', 'fare'])
dfrs
```

```python
df3=pd.concat([df3,dfrs],axis=1)
```

```python
df3.drop(['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'],axis=1,inplace=True)
df3
```

```python
df3.describe()
```

```python
# Step 5: Perform feature scaling using Power Transformer Scaler on the features: Pclass, Age, SibSp, Parch, Fare
df4=df.copy()
```

```python
dfpt= df4[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
pt=PowerTransformer()
pt
```

```python
dfpt= pt.fit_transform(dfpt)
dfpt
```

```python
dfpt= pd.DataFrame(dfpt, columns=['pclass', 'age', 'sibSp', 'parch', 'fare'])
dfpt
```

```python
df4=pd.concat([df4,dfpt],axis=1)
```

```python
df4.drop(['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'],axis=1,inplace=True)
df4
```

```python
df4.describe()
```

```python
# Step 6: Perform feature scaling using Quantile Transformer Scaler on the features: Pclass, Age, SibSp, Parch, Fare
df5=df.copy()
```

```python
dfqt= df5[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
qt=QuantileTransformer()
qt
```

```python
dfqt= qt.fit_transform(dfqt)
dfqt
```

```python
dfqt= pd.DataFrame(dfqt, columns=['pclass', 'age', 'sibSp', 'parch', 'fare'])
dfqt
```

```python
df5=pd.concat([df5,dfqt],axis=1)
```

```python
df5.drop(['Pclass', 'Age', 'SibSp', 'Parch', 'Fare'],axis=1,inplace=True)
df5
```

```python
df5.describe()
```

```python
# Concatenate all scaled dataframes with the original df
df = pd.concat([dfss, dfrs, dfms, dfpt, dfqt], axis=1)
```
