#### A.Perform different encoding operations on the dataset ‘students_adaptability_level_online_education.csv’.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

#### <span style='color:blue'>1.Perform the One Hot Encoding separately on features – Gender, Age, Internet Type, Device.</span>

```python
df=pd.read_csv(r'./students_adaptability_level_online_education.csv')
df.head()
```

```python
# Object of OneHotEncoder
ohe=OneHotEncoder()
```

```python
# One hot encoding of 'Gender'
feature_gender_array=ohe.fit_transform(df[['Gender']]).toarray()
feature_gender_array
```

```python
feature_gender_label=ohe.categories_
feature_gender_label
```

```python
gender_label=np.array(feature_gender_label).ravel()
```

```python
df_gender=pd.DataFrame(feature_gender_array,columns=gender_label,dtype=int).add_prefix('Gender_')
df_gender
```

```python
df=pd.concat([df,df_gender],axis=1)
df
```

```python
df.drop('Gender',axis=1,inplace=True)
```

```python
df
```

```python
# One hot encoding on 'Age'
feature_age_array=ohe.fit_transform(df[['Age']]).toarray()
feature_age_array
```
