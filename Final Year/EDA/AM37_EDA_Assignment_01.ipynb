```python
import pandas as pd
import numpy as np

dict1 = {
    'Emp ID': np.arange(1, 21),
    'Name': ['John Doe', 'Jane Smith', 'David Lee', 'Sarah Jones', 'Michael Brown',
             'Emily Davis', 'James Wilson', 'Jessica Garcia', 'Robert Rodriguez', 'Linda Martinez',
             'William Anderson', 'Barbara Taylor', 'Thomas Thomas', 'Elizabeth Moore', 'Christopher Jackson',
             'Ashley White', 'Matthew Harris', 'Amanda Martin', 'Kevin Thompson', 'Stephanie Garcia'],
    'Age': np.random.randint(20, 60, size=20),
    'Has Licence': np.random.choice([True, False], size=20)
}

data = pd.DataFrame(dict1)
print(data)
```
