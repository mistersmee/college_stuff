```python
historical_data.index=pd.to_datetime(historical_data.index,format='%y/%m/%d')
historical_data.index=historical_data.index.date
```

```python
historical_data
```

```python
df = historical_data[['Open','Close']]
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df.set_index('index',inplace=True)
df
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df
```

```python
yearly_max = df.resample('A').max() #year ending freq
yearly_min = df.resample('A').min()
print(yearly_max,yearly_min)
```

```python
monthly_max = df.resample('MS').max() #Monthly Starting freq
monthly_min = df.resample('MS').min()
print(monthly_max,monthly_min)
```

```python
month_end_max = df.resample('M').max() #monthly ending freq
month_end_min = df.resample('M').min()
print(month_end_max,month_end_min)
```

```python
Quaterly_max = df.resample('M').max()
Quaterly_min = df.resample('QS').min() #quarterly starting
print(Quaterly_max,Quaterly_min)
```

```python
df.resample('A')['Open'].max().plot(title='Yearly Max Open Prices', figsize=(12, 6))
plt.show()
```

```python
df.resample('MS')['Open'].max().plot(title='Monthly Max Open Prices (Begin)', figsize=(12, 6))
plt.show()
```

```python
df.resample('M')['Open'].max().plot(title='Monthly Max Open Prices (End)', figsize=(12, 6))
plt.show()
```

```python
df.resample('QS')['Open'].max().plot(title='Quarterly Max Open Prices', figsize=(12, 6))
plt.show()
```

```python
df['Close_Rolling'] = df['Close'].rolling(window=10).mean()
print(df[['Close', 'Close_Rolling']].head(15))
```

```python
df['Close'].rolling(window=10).mean().plot(title='10-Day Rolling Mean', figsize=(12, 6))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['Close'].rolling(window=10).mean().plot(label='10-Day Mean')
df['Close'].rolling(window=20).mean().plot(label='20-Day Mean')
df['Close'].rolling(window=50).mean().plot(label='50-Day Mean')
plt.legend()
plt.title('Rolling Mean (10, 20, 50 Days)')
plt.show()
```
