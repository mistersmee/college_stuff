```python
a=list(data['Country'].unique())
c= list(data['Gender'].unique())
e= list(data['Department'].unique())
data4=data.groupby('Department')
temp2=[]
for j in e:
    temp=data4.get_group(j)
    for i in a:
        b=max(temp[(temp['Country'] == i) & (temp['Gender']==c[0])]['Annual Salary'])
        d=max(temp[(temp['Country'] == i) & (temp['Gender']==c[1])]['Annual Salary'])
        temp3={'Country':i ,'Department':j ,'Gender':c[0], 'Maximum Annual Salary':b}
        temp4={'Country':i ,'Department':j, 'Gender':c[1], 'Maximum Annual Salary':d}
        temp2.append(temp3)
        temp2.append(temp4)
z=pd.DataFrame(temp2)
z = z.sort_values(by=['Country','Department'])
z
```

```python
a = list(data['Country'].unique())
c = list(data['Gender'].unique())
e = list(data['Department'].unique())
data4 = data.groupby('Department')
temp2 = []

for j in e:
    temp = data4.get_group(j)
    for i in a:
        # Check if there's any data for each gender in the country and department
        if not temp[(temp['Country'] == i) & (temp['Gender'] == c[0])].empty:
            age_c0 = min(temp[(temp['Country'] == i) & (temp['Gender'] == c[0])]['Age'])
            b = max(temp[(temp['Country'] == i) & (temp['Gender'] == c[0]) & (temp['Age'] == age_c0)]['Annual Salary'])
            temp3 = {'Country': i, 'Department': j, 'Gender': c[0], 'Age': age_c0, 'Maximum Annual Salary': b}
            temp2.append(temp3)
        
        if not temp[(temp['Country'] == i) & (temp['Gender'] == c[1])].empty:
            age_c1 = min(temp[(temp['Country'] == i) & (temp['Gender'] == c[1])]['Age'])
            d = max(temp[(temp['Country'] == i) & (temp['Gender'] == c[1]) & (temp['Age'] == age_c1)]['Annual Salary'])
            temp4 = {'Country': i, 'Department': j, 'Gender': c[1], 'Age': age_c1, 'Maximum Annual Salary': d}
            temp2.append(temp4)

z = pd.DataFrame(temp2)
z = z.sort_values(by=['Country', 'Gender'])
z
```
