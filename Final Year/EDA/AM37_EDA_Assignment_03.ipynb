```python
import numpy as np
for i in df3.columns:
    if isinstance(df3[i][0],(np.integer,np.floating)):
        df3[i].fillna(df3[i].mean(),inplace=True)
    elif not isinstance(df3[i][0],(np.integer,np.floating)):
        mode = df3[i].mode()[0]
        df3[i].fillna(mode,inplace=True)    
```
