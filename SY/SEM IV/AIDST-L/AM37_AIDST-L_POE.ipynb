```python
# Q1
import numpy as np

x = np.array([[1, 2], [3, 4]])

print("The original data is:")
print(x)

sum = np.sum(x)
print("The sum of all elements is:")
print(sum)

rsum = np.sum(x, axis = 1)
print("The sum of all elements of rows is:")
print(rsum)

csum = np.sum(x, axis = 0)
print("The sum of all elements of columns is:")
print(csum)
```

```python
# Q2
import pandas as pd

dataset = pd.read_csv("people100.csv")

df = pd.DataFrame(dataset)
df
```

```python
df2 = pd.DataFrame(df, columns = ['First Name', 'Email', 'Phone', 'Job Title'])
df2
```

```python
result = df2.loc[df['Job Title'].str.contains('Teacher|Lawyer')]
result
```

```python
df['Date of birth'] = df['Date of birth'].astype('datetime64[ns]')
df3= pd.DataFrame(df, columns = ['First Name', 'Phone', 'Date of birth', 'Job Title'])
df3
```

```python
result1 = df3.loc[df['Job Title'].str.contains('scientist|engineer')]
df4 = pd.DataFrame(result1)
df4
```

```python
result3 = df4.loc[(df4['Date of birth'] > '2000-01-01') & (df4['Date of birth'] < '2010-01-01')]
result3
```

```python
df5 = pd.DataFrame(df, columns = ['Email', 'First Name'])
result2 = df5.loc[df['First Name'].str.contains('C')]
print("Result:\n", result2.to_string())
```

```python
# Q3

data = {
    'Country': ['USA', 'India', 'China', 'Germany'],
    'Capital': ['Washington D.C.', 'New Delhi', 'Beijing', 'Berlin'],
    'Population': [331002651, 1380004385, 1444216107, 83783942],
    'No_of_states': [50, 28, 23, 16]
}


df = pd.DataFrame(data)

print("Using loc():")
print(df.loc[:, 'Country':'No_of_states'])

print("\nUsing iloc():")
print(df.iloc[:, 0:4])
```
